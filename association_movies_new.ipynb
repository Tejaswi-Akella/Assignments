```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_csv('my_movies.csv')
data.head()
```

```python
data = data.drop(['V1','V2','V3','V4','V5'],axis=1)
data
```

```python
#support value of 0.5
freq_sets_1 = apriori(data,min_support=0.5,use_colnames=True)
freq_sets_1
```

```python
results_1_conf = association_rules(freq_sets_1,metric='confidence',min_threshold=0.5)
results_1_conf
```

```python
results_1_supp = association_rules(freq_sets_1,metric='support',min_threshold=0.3)
results_1_supp
```

```python
# support value of 0.2
freq_sets_2 = apriori(data,min_support=0.2,use_colnames=True)
freq_sets_2
```

```python
results_2 = association_rules(freq_sets_2,metric='confidence',min_threshold=0.5)
results_2
```
