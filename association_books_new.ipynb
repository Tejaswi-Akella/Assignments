```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_csv('book.csv')
data.head()
```

```python
#support value of 0.1
freq_sets = apriori(data,min_support=0.1,use_colnames=True)
freq_sets
```

```python
results = association_rules(freq_sets,min_threshold=0.5,metric='confidence')
results
```

```python
# support value of 0.2
freq_sets_1=apriori(data,min_support=0.2,use_colnames=True)
freq_sets_1
```

```python
results_1 = association_rules(freq_sets_1,min_threshold=0.5,metric='confidence')
results_1
```
