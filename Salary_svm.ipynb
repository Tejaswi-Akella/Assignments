```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()

train_df['workclass'] = labelencoder.fit_transform(train_df['workclass'])
train_df['education'] = labelencoder.fit_transform(train_df['education'])
train_df['occupation'] = labelencoder.fit_transform(train_df['occupation'])
train_df['race'] = labelencoder.fit_transform(train_df['race'])
train_df['sex'] = labelencoder.fit_transform(train_df['sex'])
train_df['native'] = labelencoder.fit_transform(train_df['native'])
train_df['Salary'] = labelencoder.fit_transform(train_df['Salary'])
train_df['maritalstatus'] = labelencoder.fit_transform(train_df['maritalstatus'])
train_df['relationship'] = labelencoder.fit_transform(train_df['relationship'])
train_df
```

```python
test_df['workclass'] = labelencoder.fit_transform(test_df['workclass'])
test_df['education'] = labelencoder.fit_transform(test_df['education'])
test_df['occupation'] = labelencoder.fit_transform(test_df['occupation'])
test_df['race'] = labelencoder.fit_transform(test_df['race'])
test_df['sex'] = labelencoder.fit_transform(test_df['sex'])
test_df['native'] = labelencoder.fit_transform(test_df['native'])
test_df['Salary'] = labelencoder.fit_transform(test_df['Salary'])
test_df['relationship'] = labelencoder.fit_transform(test_df['relationship'])
test_df['maritalstatus'] = labelencoder.fit_transform(test_df['maritalstatus'])

test_df
```

```python
X_train = train_df.iloc[:,0:11]
X_train
```

```python
y_train = train_df['Salary']
y_train
```

```python
X_test = test_df.iloc[:,0:11]
X_test
```

```python
y_test = test_df['Salary']
y_test
```

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear',gamma='auto',C=3)
model = classifier.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)

from sklearn.metrics import classification_report
classification_report(y_test,y_predict)
```
