```python
from sklearn import preprocessing
labelencoder=preprocessing.LabelEncoder()
forest_df['size_category']=labelencoder.fit_transform(forest_df["size_category"])
forest_df['month']=labelencoder.fit_transform(forest_df['month'])
forest_df['day']=labelencoder.fit_transform(forest_df['day'])
forest_df
```

```python
X = forest_df.iloc[:,0:30]
X
```

```python
y = forest_df['size_category']
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

X_train.shape
```

```python
y_train.shape
```

```python
from sklearn import svm

classifier = svm.SVC(kernel='linear',gamma='auto',C=2)
```

```python
model = classifier.fit(X_train,y_train)
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_predict)
```
